```python
#1. Las ciudades ms pobladas, con los numeros de las poblaciones 
# (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). 
# (Colores: azul, gris, aceituna, marrn, con trasparencia)  
```

```python
df_ciudad_pob = df.sort_values("population_total",ascending=False)
df_ciudad_pob
```

```python
df_ciudad_pob = df_ciudad_pob[:5]
df_ciudad_pob
```

```python
barras = plt.bar(df_ciudad_pob.city, 
                df_ciudad_pob.population_total,
                color = ["blue","grey","orange","olive","brown"], 
                alpha=0.5)
plt.title("Ciudades ms pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)")
#plt.text(0,0, s=df.ciudad_pob.city)
plt.subplots_adjust(wspace = 0.5)

for i, valor in enumerate(df_ciudad_pob.population_total):
    plt.text(i,valor,
        f"{valor}",
        ha="center", va="bottom"
    )

plt.tight_layout()
plt.show()
```

```python
#2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
```

```python
plt.hist(df_california["area_total_km2"],bins=60 ,alpha= 0.5, color="steelblue",label="Hist area")
plt.xlabel("Superficie (km2)",fontdict={"family":"MS Gothic"})
plt.ylabel("Nmero de ciudades",fontdict={"family":"MS Gothic"})
plt.legend(loc="upper right")
plt.tight_layout()
plt.show();
```

```python
#3. El scatter de rea de agua de un ciudad con su poblacion.
```

```python
plt.scatter(df_california["area_water_km2"], df_california["population_total"]/1000,c= "g", alpha=0.5)
plt.xlabel("rea de agua en la ciudad (km2)")
plt.ylabel("Poblacin (miles habitantes)")
plt.title("Poblacin por kilmetro cuadrado de agua")
#primera flecha
plt.annotate("",
    xy=(120, 1400),
    xytext=(55, 2900),
    arrowprops=dict(facecolor="black", arrowstyle="->"),
    fontsize=10, family="serif")
#segunda flecha
plt.annotate(
    "Outliers",
    xy=(80, 3500),
    xytext=(40, 2800),
    arrowprops=dict(facecolor="black", arrowstyle="->"),
    fontsize=10, family="serif")

plt.tight_layout()
plt.show();
```

```python
#el primer grafioc
fig, axs = plt.subplots(2, 2, figsize=(16, 16))
axs[0,0].bar(df_ciudad_pob.city, 
                df_ciudad_pob.population_total,
                color = ["blue","grey","orange","olive","brown"], 
                alpha=0.5)
axs[0,0].set_title("Ciudades ms pobladas de California")
axs[0,0].set_xlabel("Ciudades")
axs[0,0].set_ylabel("Habitantes (miles)")

for i, valor in enumerate(df_ciudad_pob.population_total):
    axs[0,0].text(
        i, valor,
        f"{valor}",
        ha="center", va="bottom")
#segundo
axs[0,1].hist(df_california["area_total_km2"],bins=60 ,alpha= 0.5, color="steelblue",label="Hist area")
axs[0,1].set_xlabel("Superficie (km2)",fontdict={"family":"MS Gothic"})
axs[0,1].set_ylabel("Nmero de ciudades",fontdict={"family":"MS Gothic"})
axs[0,1].legend(loc="upper right")

#tercero
axs[1,0].scatter(df["area_water_km2"], df["population_total"]/1000, c="g",alpha=0.3)
axs[1,0].set_xlabel("rea de agua en la ciudad (km2)")
axs[1,0].set_ylabel("Poblacin (miles habitantes)")
axs[1,0].set_title("Poblacin por kilmetro cuadrado de agua")
axs[1,0].annotate(
    "", 
    xy=(120, 1400), 
    xytext=(55, 2900),
    arrowprops=dict(facecolor="black", arrowstyle="->"),
    fontsize=10,family="serif")
axs[1,0].annotate(
    "Outliers", 
    xy=(80, 3500), 
    xytext=(40, 2800),
    arrowprops=dict(facecolor="black", arrowstyle="->"),
    fontsize=10, family="serif")

#cuarto
axs[1,1].boxplot(df["elevation_m"], vert=False, patch_artist=True, boxprops=dict(facecolor='red',color='black'))
axs[1,1].set_title("Diagrama de caja de la Altitud de las ciudades de California")
axs[1,1].set_xlabel("Altitud (m)")


plt.tight_layout()
plt.show()
```
