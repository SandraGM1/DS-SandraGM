## Métrica:
El error cuadrático medio (RMSE, por sus siglas en inglés) es una medida de la desviación estándar de los residuos (errores de predicción). Los residuos representan la diferencia entre los valores observados y los valores predichos por el modelo. El RMSE indica qué tan dispersos están estos errores: cuanto menor es el RMSE, más cercanas están las predicciones a los valores reales. En otras palabras, el RMSE mide qué tan bien se ajusta la línea de regresión a los datos.


$$ RMSE = \sqrt{\frac{1}{n}\Sigma_{i=1}^{n}{\Big(\frac{d_i -f_i}{\sigma_i}\Big)^2}}$$

```python
# Categóricas
cat_pipeline = Pipeline([
    ("Impute_Mode", SimpleImputer(strategy="most_frequent")),
    ("OHEncoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

# Numéricas con log
num_pipeline_log = Pipeline([
    ("Impute_Mean", SimpleImputer(strategy="median")),
    ("Apply_Logarithm", FunctionTransformer(np.log1p, feature_names_out='one-to-one')),
    ("SScaler", StandardScaler())
])

# Numéricas sin log
num_pipeline = Pipeline([
    ("Impute_Mean", SimpleImputer(strategy="median")),
    ("SScaler", StandardScaler())
])
```

```python
preprocessing = ColumnTransformer(
    [
        ("Process_Numeric_Log", num_pipeline_log, num_log_cols),
        ("Process_Numeric", num_pipeline, num_no_log_cols),
        ("Process_Categorical", cat_pipeline, categoricas),
        ("Exclude", "drop", columns_to_exclude)
    ],
    remainder="passthrough"
)
```

## 4. Modelado

### 4.1 Baseline de modelos


```python
pipeline_preprocesado = Pipeline([
    ("Mega", FunctionTransformer(tdf.transformar_dataset, validate=False)),
    ("Prep", preprocessing)
])
```

```python
pipe_rf = Pipeline([
    ("Pre", pipeline_preprocesado),
    ("Model", RandomForestRegressor(random_state=42, n_jobs=-1))
])
```

```python
pipe_gb = Pipeline([
    ("Pre", pipeline_preprocesado),
    ("Model", GradientBoostingRegressor())
])
```

```python
pipe_xgb = Pipeline([
    ("Pre", pipeline_preprocesado),
    ("Model", XGBRegressor(random_state=42, n_jobs=-1, tree_method="hist"))
])
```

```python
pipe_lr = Pipeline([
    ("Pre", pipeline_preprocesado),
    ("Model", LinearRegression())
])
```

```python
pipe_ridge = Pipeline([
    ("Pre", pipeline_preprocesado),
    ("Model", Ridge())
])
```

```python
pipe_lasso = Pipeline([
    ("Pre", pipeline_preprocesado),
    ("Model", Lasso(max_iter=10000))
])
```

```python
pipe_elastic = Pipeline([
    ("Pre", pipeline_preprocesado),
    ("Model", ElasticNet(max_iter=10000))
])

```

```python
pipe_knn = Pipeline([
    ("Pre", pipeline_preprocesado),
    ("Model", KNeighborsRegressor())
])
```

```python
pipe_et = Pipeline([
    ("Pre", pipeline_preprocesado),
    ("Model", ExtraTreesRegressor(random_state=42, n_jobs=-1))
])
```

```python
modelos_baseline = {
    "RandomForest": pipe_rf,
    "GradientBoosting": pipe_gb,
    "XGBoost": pipe_xgb,
    "LinearRegression": pipe_lr,
    "Ridge": pipe_ridge,
    "Lasso": pipe_lasso,
    "ElasticNet": pipe_elastic,
    "KNN": pipe_knn,
    "ExtraTrees": pipe_et
}
```

### 4.2 Sacar métricas, valorar los modelos

Recuerda que en la competición se va a evaluar con la métrica de ``RMSE``.

```python
from sklearn.metrics import mean_squared_error
import numpy as np

resultados = {}

for nombre, pipe in modelos_baseline.items():
    print(f"Entrenando {nombre}...")
    
    pipe.fit(X_train_raw, y_train)
    pred = pipe.predict(X_test_raw)
    
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    resultados[nombre] = rmse
    
    print(f"{nombre}: RMSE = {rmse:.4f}")
```

```python
for nombre, rmse in resultados.items():
    print(f"{nombre} {rmse:.4f}")
#baseline
```
